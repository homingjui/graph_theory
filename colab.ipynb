```python
from itertools import product
import numpy as np
```

```python
num = 6
dots = 6
L=[[1,2],[1,3],[1,4],[2,3],[2,4],[3,4]]

print_str=""
for i in L:
  print_str+="L("+str(i[0])+","+str(i[1])+";%d) "

iter = product(np.arange(num),repeat=num)
for g in iter:
  if sum(g)==dots:
    print(print_str%g)
    graph=[]
    for item,index in enumerate(g,start=1):
      if index==0:
        graph.append([L[item-1][0],L[item-1][1]])
      else:
        for i in range(index+1):
          graph.append([L[item-1][0],L[item-1][0]*10+L[item-1][1]+(i+1)/10])
          graph.append([0,L[item-1][0]*10+L[item-1][1]+(i+1)/10])
    print(graph)
```
